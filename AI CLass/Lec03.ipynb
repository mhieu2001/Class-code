    - per capita crime rate by town

  - LSTAT
    - % lower status of the population

  - MEDV
    - Median value of owner-occupied homes in $1000's

# Linear regression(線形回帰分析)
- 説明変数$x_i$に対して目的変数$y_i$ $(i=1,2,\cdots, n)$が一次(線形)式で表現すること。
- The objective variable $y_i$ $(i=1,2,\cdots, n)$ for the explanatory variable $x_i$ should be expressed by a linear (linear) expression.
  - 一次式とは$y=f(x)=ax+b$となる
  - A linear equation is $y=f(x)=ax+b$.

## Least-squares method(最小二乗法)
- データ$(x_i, y_i)$  $(i=1,2,\cdots, n)$ が与えられたとき、$x_i$ と $y_i$ の関係に対してもっともらしい関数$y=f(x)$ を求める方法
- Given data $(x_i, y_i)$  $(i=1,2,\cdots, n)$, a method to find a plausible function $y=f(x)$ for the relationship between $x_i$ and $y_i$.
- E.g)
  - 図中の丸がデータ、オレンジの直線が求めようとしている関数$y=f(x)=ax+b$としたときに、最小二乗法は、各データと求めようとしている関数の間の最小の距離の和(最小二乗和)が最小になるように関数を決定する手法
  - When the circles in the figure represent data and the orange line represents the function $y=f(x)=ax+b$, the least-squares method determines the function so that the sum of the smallest distances (sum of least squares) between each data and the function to be obtained is minimized.

image.png

### Formulas for functions by the method of least squares in the case of linear regression
- データ$(𝑥𝑖,𝑦𝑖)$  $(𝑖=1,2,⋯,𝑛)$が与えられた際に、$𝑥$と$𝑦$の関係をもっともらしい関数$𝑦=f(x)=ax+b$の$a,b$は次の式で表される
- Given data $(𝑥𝑖,𝑦𝑖)$ $(𝑖=1,2,⋯,𝑛)$, the $a,b$ of the function $𝑦=f(x)=ax+b$ that plausibly relates $𝑥$ and $𝑦$ can be expressed as follows

  - $a=\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{\sum_{i=1}^n(x_i-\bar{x})^2}=\frac{s_{xy}}{\sigma_x^2}$
  - $b=\bar{y}-a\bar{x}$
    - $\bar{x}$: $x_i$の平均値(Average value of $x_i$.)
    - $\bar{y}$: $y_i$の平均値(Average value of $y_i$.)
    - $\sigma_x^2$: $x_i$の分散(Variance of $x_i$.)
    - $s_{xy}$: $x_i,y_i$の共分散(Covariance of $x_i,y_i$.)


## How to find the least-squares formula in the case of linear regression
- Data $(x_i,y_i)$ $(i=1,2,\cdots, n)$
- Function to be found: $y=f(x)=ax+b$
  - In other words, we want to find $a,b$.
- Find the least-squares sum of each data $(x_i,y_i)$ and the function ($y=f(x)=ax+b$)
  - $\sum_{i=1}^n \{y_i-(ax_i+b)\}^2$
  -
- データ$(x_i,y_i)$  $(i=1,2,\cdots, n)$
- 求めたい関数：$y=f(x)=ax+b$
  - つまり$a,b$を求めていく
- 各データ$(x_i,y_i)$と関数($y=f(x)=ax+b$)との最小二乗和を求める
  - $\sum_{i=1}^n \{y_i-(ax_i+b)\}^2$


```python
import sympy
sympy.var('a,b,x,y')
J=sympy.expand((y-(a*x+b))**2)
J
```

- $a^2\sum_{i=1}^nx_i^2+2ab\sum_{i=1}^nx_i-2a\sum_{i=1}^nx_iy_i+nb^2-2b\sum_{i=1}^ny_i+\sum_{i=1}^ny_i^2 $
  - 注意：$\sum_{i=1}^nb=nb$
- 上記の式をa, bについて偏微分(Partial differentiation of the above equation for a, b)


```python
dJda=sympy.diff(J,a)
dJda
```

```python
dJdb=sympy.diff(J,b)
dJdb
```

- $2a \sum_{i=1}^nx_i^2 +2b \sum_{i=1}^nx_i-2 \sum_{i=1}^nx_iy_i=0 $　　①
- $2a \sum_{i=1}^nx_i+2nb-2\sum_{i=1}^ny_i=0$　　②

- ②式を$b$についての式に展開
  - $2nb-=2\sum_{i=1}^ny_i-2a \sum_{i=1}^nx_i$
  - $nb=\sum_{i=1}^ny_i-a \sum_{i=1}^nx_i$
  - $b=\frac{\sum_{i=1}^ny_i}{n}-a\frac{\sum_{i=1}^nx_i}{n}$
    - $\frac{\sum_{i=1}^nx_i}{n}$は、データ$x_i$の平均を意味するため$\bar{x}$とおき、$\frac{\sum_{i=1}^ny_i}{n}$はデータ$y_i$の平均を意味するため$\bar{y}$とおく
  - $b=\bar{y}-a\bar{x}$　　②'

- ②'式を①式に代入し、$a$についての式に展開
  - $2a \sum_{i=1}^nx_i^2 +2(\bar{y}-a\bar{x}) \sum_{i=1}^nx_i-2 \sum_{i=1}^nx_iy_i=0 $
    - 両編に2を割る
  - $a\sum_{i=1}^nx_i^2+(\bar{y}-a\bar{x}) \sum_{i=1}^nx_i-\sum_{i=1}^nx_iy_i=0$
  - $a\sum_{i=1}^nx_i^2+\bar{y}\sum_{i=1}^nx_i-a\bar{x}\sum_{i=1}^nx_i -\sum_{i=1}^nx_iy_i=0$
  - $a(\sum_{i=1}^nx_i^2-\bar{x}\sum_{i=1}^nx_i)=\sum_{i=1}^nx_iy_i-\bar{y}\sum_{i=1}^nx_i$
  - $a=\frac{\sum_{i=1}^nx_iy_i-\bar{y}\sum_{i=1}^nx_i}{\sum_{i=1}^nx_i^2-\bar{x}\sum_{i=1}^nx_i}$
  -
  - $a=\frac{\frac{\sum_{i=1}^nx_iy_i}{n}-\bar{xy}}{\frac{\sum_{i=1}^nx_i^2}{n}-\bar{x}^2}$
    -
    - $\frac{\sum_{i=1}^nx_i^2}{n}-\bar{x}^2=\frac{\sum_{i=1}^nx_i^2}{n}-2\bar{x}^2+\bar{x}^2=\frac{1}{n}(\sum_{i=1}^nx_i^2-2\bar{x}\frac{\sum_{i=1}^nx_i}{n}+n\bar{x}^2)=\frac{1}{n}\sum_{i=1}^n(x_i-\bar{x})^2$
    - $\frac{\sum_{i=1}^nx_iy_i}{n}-\bar{xy}=\frac{\sum_{i=1}^nx_iy_i}{n}-2\bar{xy}+\bar{xy}=\frac{1}{n}(\sum_{i=1}^nx_iy_i-2n\bar{xy}+n\bar{xy})=\frac{1}{n}(\sum_{i=1}^nx_iy_i-n\frac{\sum_{i=1}^nx_i}{n}\bar{y}-n\bar{x}\frac{\sum_{i=1}^ny_i}{n}+n\bar{xy})=\frac{1}{n}\sum_{i=1}^n (x_iy_i-x_i\bar{y}-\bar{x}y_i+\bar{xy})=\frac{1}{n}\sum_{i=1}^n{(x_i-\bar{x})(y_i-\bar{y})}$
  -
  - $a=\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{\sum_{i=1}^n(x_i-\bar{x})^2}$

    - $\sum_{i=1}^n(x_i-\bar{x})^2$は、$x$の分散と呼ばれ$\sigma_x^2$と表現($\sigma_x$は標準偏差)
    - $\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})$は$x,y$の共分散と呼ばれ$s_{xy}$(もしくは$Cov[X,Y]$)と表現

## Linear regression using 'lstat' and 'medv' from Boston Housing Data Set
- LSTAT（低所得者の割合)が高くなるほどMEDV（住宅価格の中央値）低くなるのではないかと仮定
- 説明変数$x$をLSTAT、目的変数$y$をMEDV(住宅価格の中央値)として線形回帰を実装してみる
- Assume that the higher the LSTAT (percentage of low-income residents), the lower the MEDV (median home price)
- Implement linear regression with $x$ as the explanatory variable LSTAT and $y$ as the objective variable MEDV (median housing price)

### Data Preparation

```python
df
```

```python
x=df['lstat'].values
y=df['medv'].values
```

### Calculate according to the least-squares formula
- $a=\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{\sum_{i=1}^n(x_i-\bar{x})^2}=\frac{s_{xy}}{\sigma_x^2}$
- $b=\bar{y}-a\bar{x}$

```python
x.mean()
```

```python
y.mean()
```

```python
a=np.sum((x-x.mean())*(y-y.mean()))/np.sum((x-x.mean())**2)
a
```

```python
b=y.mean()-a*x.mean()
b
```

### Draw a graph of the results of a linear regression

```python
import numpy as np
x_input=np.linspace(0, 40, 10000)
x_input
```

```python
y_output=a*x_input+b
y_output
```

```python
plt.scatter(x,y)
plt.plot(x_input,y_output, color='orange')

plt.xlabel('lstat')
plt.ylabel('medv')
plt.show()
```

## Evaluation index for Regression (回帰モデルの評価手法)

```python
y_hat=a*x+b
y_hat
```

### Mean Squared Error(MSE, 平均二乗誤差)
- データ$(x_i, y_i)$、回帰式による推定値$\hat{y}_i$
  - $MSE=\frac{1}{n}\sum_{i=1}^n (y_i-\hat{y}_i)^2$

```python
y_hat.shape[0]
```

```python
n=y_hat.shape[0]
(1/n)*np.sum((y-y_hat)**2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_hat)
```

### Root Mean Squared Error (RMSE, 平均平方二乗誤差)
- データ$(x_i, y_i)$、回帰式による推定値$\hat{y}_i$
  - $RMSE=\sqrt{\frac{1}{n}\sum_{i=1}^n (y_i-\hat{y}_i)^2}$

```python
np.sqrt((1/n)*np.sum((y-y_hat)**2))
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y, y_hat))
```

### Mean Absolute Error (MAE, 平均絶対誤差)
- データ$(x_i, y_i)$、回帰式による推定値$\hat{y}_i$
  - $RMSE=\frac{1}{n}\sum_{i=1}^n |y_i-\hat{y}_i|$

```python
(1/n)*np.sum(np.abs(y - y_hat))
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y, y_hat)
```

### Coefficient of Determination(R2, R squared, 決定係数、寄与率)
- 決定係数は、回帰式が元データ(正解データ)にどれくらい当てはまるかを0～1.0で表現し、1.0に近いほど、回帰式が実際のデータに当てはまっていることをを示す
- The coefficient of determination expresses how well the regression equation model fits the correct data on a scale of 0 to 1.0, with the closer to 1.0 indicating that the regression equation fits the actual data

#### General Definition
- データ$(x_i, y_i)$、回帰式による推定値$\hat{y}_i$
  - $R^2=1-\frac{\sum_{i=1}^n (y_i-\hat{y}_i)^2}{\sum_{i=1}^n (y_i-\bar{y})^2}$

```python
r2=1-np.sum((y-y_hat)**2)/np.sum((y-y.mean())**2)
r2
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_hat)
```

#### Coefficient of determination adjusted for degrees of freedom (自由度調整済みの決定係数)
- データ$(x_i, y_i)$、回帰式による推定値$\hat{y}_i$、説明変数の数をp、データの数をn

  - $R^2=1-\frac{\frac{1}{n-p-1}\sum_{i=1}^n (y_i-\hat{y}_i)^2}{\frac{1}{n-1}\sum_{i=1}^n (y_i-\bar{y})^2}$

```python
n=y_hat.shape[0]
n
```

```python
p=1
```

```python
r2=1-(1/(n-p-1)*np.sum((y-y_hat)**2))/(1/(n-1)*np.sum((y-y.mean())**2))
r2
```

## Change in order and coefficient of determination R2(次数の変化と決定係数R2の変化)
- 6次多項式回帰まで決定係数R2が上がり、それ以降は決定係数R2下がる-->過学習(Overfitting)
- The coefficient of determination R2 increases until 6th order polynomial regression, after which the coefficient of determination R2 decreases-->Overfitting

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

order=np.array([])
R2_train=np.array([])
R2_test=np.array([])

for i in range(13):
  if i == 0:
    pass
  if i > 0:
    if i == 1:
      X_train_poly = X_train
      X_test_poly = X_test
    else:
      poly = PolynomialFeatures(i)
      X_train_poly = poly.fit_transform(X_train)
      X_test_poly = poly.fit_transform(X_test)

    model = LinearRegression()
    model.fit(X_train_poly, y_train)
    y_train_hat=model.predict(X_train_poly)
    y_test_hat=model.predict(X_test_poly)

    order=np.append(order,i)
    R2_train = np.append(R2_train, r2_score(y_train, y_train_hat))
    R2_test = np.append(R2_test, r2_score(y_test, y_test_hat))


plt.plot(order,R2_train, color='blue', label='r2_train')
plt.plot(order,R2_test, color='orange', label='r2_test')

plt.legend()
plt.xlabel('order')
plt.ylabel('R2')
plt.show()
```

### To prevent overfitting(過学習を防ぐ方法)
- データを増やす
  - data augmentation
- 単純なモデルへ変更する
- L1, L2正則化(Normalization of L1 or L2)
  - モデルの複雑性と学習データへの当てはまりの良さをバランスさせる手法
  - 過学習を起こすとそれぞれの係数の絶対値が非常に大きくなるため、それを小さくなるように調整
  - A method to balance the complexity of the model with the fit to the training data.
  - Over-training leads to very large absolute values of the respective coefficients, which can be reduced.
- 訓練回数の早期停止(Early stopping of the number of training cycles)
  - Dropout：
    - 事前に割合に応じて、ランダムにノードを選択し、そのノードの重みを0
    -  Randomly select a node according to a prior proportion and set its weight to 0
  - DropConnect
    - 上記のDropoutの0の代わりに任意の重みに設定
    - Set to any weight instead of 0 for Dropout above

### Ridge regression(Ridge回帰)
- 損失関数に回帰係数の2乗和(L2正則化項)を導入した手法
- A method that introduces the sum of squares of the regression coefficients (L2 regularization term) into the loss function
-
- 通常の最小二乗法で回帰係数を求めるときの損失関数
  - $\sum_{i=1}^n (y_i - \hat{y}_i)^2$
    - $\hat{y}_i=\theta_0+\theta_1x_i+\theta_2x_i^2+\cdots\theta_px_i^p$
- Ridge回帰での回帰係数を求めるときの損失関数
  - $\sum_{i=1}^n (y_i - \hat{y}_i)^2+\frac{1}{2}\alpha\sum_{j=1}^p\theta_k^2$
    - $\hat{y}_i=\theta_0+\theta_1x_i+\theta_2x_i^2+\cdots\theta_px_i^p$
    - $ \alpha$: 分析者が決定するパラメータ

-
- この例では$p=11$

```python
from sklearn.linear_model import Ridge

poly = PolynomialFeatures(11)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

model = Ridge()
model.fit(X_train_poly, y_train)

y_train_hat=model.predict(X_train_poly)
y_test_hat=model.predict(X_test_poly)

r2_train=r2_score(y_train, y_train_hat)
r2_test=r2_score(y_test, y_test_hat)

print('coef:', model.coef_)
print('R2:', r2_train, r2_test)

x_input=np.linspace(0, 37, 10000).reshape(-1, 1)
x_input_poly = poly.fit_transform(x_input)
y_output=model.predict(x_input_poly)

plt.scatter(x,y)
plt.plot(x_input,y_output, color='orange')
```

### Lasso regression(Lasso回帰)
- 損失関数に回帰係数の絶対値の和(L1正則化項)を導入した手法
- A method that introduces the sum of the absolute values of the regression coefficients (L1 regularization term) into the loss function
-
- 通常の最小二乗法で回帰係数を求めるときの損失関数
  - $\sum_{i=1}^n (y_i - \hat{y}_i)^2$
    - $\hat{y}_i=\theta_0+\theta_1x_i+\theta_2x_i^2+\cdots\theta_px_i^p$
- Lasso回帰での回帰係数を求めるときの損失関数
  - $\sum_{i=1}^n (y_i - \hat{y}_i)^2+\frac{1}{2}\alpha\sum_{j=1}^p|\theta_k|$
    - $\hat{y}_i=\theta_0+\theta_1x_i+\theta_2x_i^2+\cdots\theta_px_i^p$
    - $ \alpha$: 分析者が決定するパラメータ

-
- この例では$p=11$

```python
from sklearn.linear_model import Lasso

poly = PolynomialFeatures(11)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

model = Lasso()
model.fit(X_train_poly, y_train)

y_train_hat=model.predict(X_train_poly)
y_test_hat=model.predict(X_test_poly)

r2_train=r2_score(y_train, y_train_hat)
r2_test=r2_score(y_test, y_test_hat)

print('coef:', model.coef_)
print('R2:', r2_train, r2_test)

x_input=np.linspace(0, 37, 10000).reshape(-1, 1)
x_input_poly = poly.fit_transform(x_input)
y_output=model.predict(x_input_poly)

plt.scatter(x,y)
plt.plot(x_input,y_output, color='orange')
```

- 使いわけメモ
  - Ridge回帰をまず試す
    - Ridgeは全体的に回帰変数を小さくする
    - Lassoは特徴選択をしながら回帰変数を小さくする
  - 特徴量が多く、そのうち重要なものは僅かしかないことが予測されるのであればLasso
-
- Usage notes
  - Try Ridge regression first
    - Ridge reduces regression variables overall
    - Lasso reduces regression variables while selecting features
  - If you have a large number of features, of which only a few are expected to be important, use Lasso.

- Ridge回帰とLasso回帰の折衷案
- Elastic Netでの回帰係数を求めるときの損失関数
  - $\sum_{i=1}^n (y_i - \hat{y}_i)^2+\lambda_1\sum_{j=1}^p\theta_k^2+\lambda_2\sum_{j=1}^p|\theta_k|$
    - $\hat{y}_i=\theta_0+\theta_1x_i+\theta_2x_i^2+\cdots\theta_px_i^p$

## Support Vector Regression
- Support Vector Machines (SVMs) applied to regression analysis
- Algorithm that implements classification based on the concept of margin maximization.
  - The algorithm tries to obtain more reasonable boundaries by increasing the margin between the decision boundary and the training data.
-
- Details are given in Support Vector Machine (Classification Problem)
-

- SVRでは𝜀-不感損失関数を利用する
- 損失関数:$\epsilon_i=y_i-f(x_i)$
- SVRは損失関数を計算するが、誤差がある一定値$\epsilon$を超えるまでは0と扱う𝜀-不感損失関数を使う。$−\epsilon$<誤差<$\epsilon$の範囲のデータ点は誤差評価の対象外
- 非線形のカーネル関数（RBFカーネル、多項式カーネル、シグモイドカーネル等）を使うことで非線形の回帰分析を行うことも可

-
- SVR uses 𝜀-insensitive loss function
- Loss function:$\epsilon_i=y_i-f(x_i)$
- SVR calculates the loss function, but uses an 𝜀-insensitive loss function that treats the error as zero until it exceeds a certain value of $\epsilon$. Data points in the range of $-\epsilon$<error<$\epsilon$ are not subject to error evaluation.
- Non-linear regression analysis can also be performed by using non-linear kernel functions (RBF kernel, polynomial kernel, sigmoid kernel, etc.)



```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd

model = SVR(C=1.0, kernel='rbf', epsilon=0.1, gamma='auto')    # RBF kernel
model.fit(X_train, y_train)
y_pred=model.predict(X_test)

# evaluation
# R2
r2 = r2_score(y_test, y_pred)

# MAE
mae = mean_absolute_error(y_test, y_pred)

# RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("R2 : %.3f" % r2)
print("MAE : %.3f" % mae)
print("RMSE : %.3f" % rmse)
```
