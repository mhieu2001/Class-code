## Retrieve tabular data from the web and graph it.
## Get a table on the web
- The <Table> tag part of HTML can be parsed and stored in a Dataframe
- e.g.) Monthly average temperature values for Tokyo (°C)
  - https://www.data.jma.go.jp/obd/stats/etrn/view/monthly_s3.php?prec_no=44&block_no=47662

```python
import pandas as pd

url = 'https://www.data.jma.go.jp/obd/stats/etrn/view/monthly_s3.php?prec_no=44&block_no=47662'
dfs = pd.read_html(url)
dfs
```

## Align the values in the table with the values of type float64 (floating point type).
- The actual contents of the df contain values such as "22.6 )", "15.7 ]" and "NaN".
  - 'NaN' is a real missing value if there is no value
  - Values with a ")", such as "22.6 )", indicate that some days could not be measured, but the value is statistically acceptable for taking monthly or yearly averages.
  - A missing value with "]", such as "15.7 ]", indicates that some days could not be measured and there is not enough statistical data to take a monthly or yearly average, but the average value was forced.
- In this case, it is sufficient to be able to identify a trend, so values with ")" and "]" are also treated as normal averages.
- For this purpose, it is necessary to remove the ")" and "]".
  - Note that "22.6 )" and "15.7 ]" contain a space between the numerical value and the symbols ")" and ")".
- The policy is to align all values to type string, eliminate the spaces and the symbols ")" and "]", and then align all values except the missing value represented by "NaN" to type float64 (floating point type).

```python
df.info()
```

- df.info()でdfの中の値がどういう型の値なのか確認。
- float64かobject型となっている
- 「]」、「)」、空白を削除するためには、一旦string型(文字列型)に揃えてから、削除を実行し、さらに、float64型(浮動小数点型)に揃えるという方法をとらなければならない

- 途中、string型(文字列型)からfloat64型(浮動小数点型)に型を変更する必要があるが、一部欠損値「NaN」がまざっているため、string型(文字列型)からfloat64型(浮動小数点型)に型を変更が成功すればそのままfloat64型(浮動小数点型)に直し、エラーとなればNaN(欠損値)を割り当てるという関数を作っておく

```python
import numpy as np

def str2float(s):
    try:
        return np.float64(s)
    except:
        return np.nan
```

- df.columns.valuesは列名を配列で取得する。
- for t in df.columns.values:でtに「1月」から順番に代入しながら繰り返すという意味。
- df[t]=df[t].astype('str')はstring型(文字列型)に一旦変更。
- df[t]=df[t].str.strip(']')は「]」が見つかれば削除
- df[t]=df[t].str.strip(')')は「)」が見つかれば削除
- df[t]=df[t].str.strip()は空白が見つかれば削除
- df[t]=str2float(df[t])は上記で定義したstr2floatを使ってstring型(文字列型)からfloat64型(浮動小数店型)に変換

```python
df.columns.values
```

```python
for t in df.columns.values:
  df[t]=df[t].astype('str')
  df[t]=df[t].str.strip(']')
  df[t]=df[t].str.strip(')')
  df[t]=df[t].str.strip()
  df[t]=str2float(df[t])
df
```
